```python
import pandas as pd
```

```python
data = pd.read_csv("resources/budget_data.csv")
```

```python
data
```

```python
# Count number of months in data set
months = data["Date"].count()
months
```

```python
#Total profits and losses
revenue = data["Revenue"].sum()
```

```python
#The average change in revenue between months over the entire period
avg_revenue = data["Revenue"].mean()
```

```python
#The greatest increase in revenue
max_revenue = data["Revenue"].max()
max_revenue

max_date = data.loc[data["Revenue"]==max_revenue, "Date"]
```

```python
#The greatest increase in revenue
min_revenue = data["Revenue"].min()
min_revenue

min_date = data.loc[data["Revenue"]==min_revenue, "Date"]
```
